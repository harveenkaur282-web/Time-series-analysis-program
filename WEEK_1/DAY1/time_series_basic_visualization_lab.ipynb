# Lab: What is Time Series (Basic Visualization)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np

data = {
    "date": pd.date_range(start="2024-01-01", periods=30, freq='D'),
    "value": np.arange(10, 40) + np.random.randint(-2, 3, 30)
}
```

```python
df = pd.DataFrame(data)
```

```python
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df['value'], marker='o')
plt.title("Basic Time Series Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid()
plt.show()
```

## Observations
- The data shows an **increasing trend** over time.
- There are small fluctuations (noise).
- Time (date) is essential to understand the pattern.

This is the basic idea of **time series analysis**.
